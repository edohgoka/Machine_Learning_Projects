**GOAL**: this project is to apply the **PCA** technique on data to see how it behaves to find new features that may be useful for dimensionality reduction and to apply other machine learning algorithms if necessary thereafter.

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from pca import pca
import pandas as pd
```

```python
df = pd.read_excel("data_vacations_review.xlsx")
```

```python
df
```

```python
df.columns
```

```python
df.set_index(["name"])
```

```python
X = df.reset_index(drop=True).drop(["name"], axis=1)
```

```python
X
```

```python
# reduce the data towards 2 PCs
model = pca(n_components=2)
```

```python
# Fit transform
results = model.fit_transform(X)
```

```python
# Plot explained variance
fig, ax = model.plot()
```

```python
# Scatter first 2 PCs
fig, ax = model.scatter()
plt.show()
```

```python
# Make biplot with the number of features
fig, ax = model.biplot(n_feat=4, y = df["name"].values)
```

# Conclusion 

In this project, we took a data of some people rating about the types of vacations they like. By applying the **PCA** on the data, we have created new features to cluster the preferences of people. It was shown that the the first PCA component helps us to differentiate pepople who like a **relax** vacation from those who like an **adventure** one. The second PCA component helps us to categorize people who like vacations in **montain** from those who like them nearby the **sea**. It is important to point out that the two components of PCA have almost 100% of variance explained.
